## Arithematic Operations
### Addition

```python
with pm.Model() as model:
    x = RV(tfd.Normal(0., 1.), name="x", sample_shape=10000)
    y = RV(tfd.Normal(1., 1.), name="y", sample_shape=10000)
    z = x + y
with tf.Session() as sess:
    sns.distplot(sess.run(z))
    plt.plot()
    sns.distplot(sess.run(x))
    plt.plot()
    sns.distplot(sess.run(y))
    plt.plot()
```

### Multiplication

```python
with pm.Model() as model:
    x = RV(tfd.Normal(0., 1.), name="x", sample_shape=10000)
    y = RV(tfd.Normal(1., 1.), name="y", sample_shape=10000)
    z = x * y
with tf.Session() as sess:
    sns.distplot(sess.run(z))
    plt.plot()
    sns.distplot(sess.run(x))
    plt.plot()
    sns.distplot(sess.run(y))
    plt.plot()
```

### Substraction

```python
with pm.Model() as model:
    x = RV(tfd.Normal(0., 1.), name="x", sample_shape=10000)
    y = RV(tfd.Normal(1., 1.), name="y", sample_shape=10000)
    z = x - y
with tf.Session() as sess:
    sns.distplot(sess.run(z))
    plt.plot()
    sns.distplot(sess.run(x))
    plt.plot()
    sns.distplot(sess.run(y))
    plt.plot()
```
